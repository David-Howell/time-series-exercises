## 3. Extract the data for [sales](https://python.zgulde.net/api/v1/sales). There are a lot of pages of data here, so your code will need to be a little more complex. Your code should continue fetching data from the next page until all of the data is extracted.

```python
url_sales = 'https://python.zgulde.net/api/v1/sales'
```

```python
requests.get(url_sales).json()
```

```python
sales_response = requests.get(url_sales).json()['payload']
```

```python
sales_response
```

```python
ir = sales_response

sales = []

ir_n = ir
for i in range(ir['max_page']):

    sales.append(ir_n['sales'])

    if ir_n['page'] == i+1: break

    next_page = ir['next_page']

    ir_n = requests.get('https://python.zgulde.net' + next_page).json()

# items = pd.DataFrame(items)

for i in sales:
    sales = i

sales = pd.DataFrame(sales)

sales
```
